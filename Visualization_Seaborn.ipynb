## Distribution Plots

### a. distribution plot
<li> Provides a way to look at a univariate distribution. </li>
<li> A univeriate distribution provides a distribution for one variable</li>
<li> Kernal Density Estimation with a Histogram is provided</li>
<li> kde=False removes the KDE</li>
<li> Bins define how many buckets to divide the data up into between intervals</li>
<li> For example put all profits between 10 and 20$ in this bucket</li>


```python
sns.distplot(tips['total_bill'],
            bins = 10,
            kde = True)
plt.show()
```

```python
sns.distplot(tips['total_bill'],
            bins = 10,
            kde = False)
plt.show()
```

```python
sns.kdeplot(tips['total_bill'])
plt.show()
```

<li> To create a heatmap with data you must have data set up as a matrix where variables
are on the columns and rows

<li> Correlation tells you how influential a variable is on the result
<li> Colormaps - https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html

```python
car_df = sns.load_dataset('car_crashes')
car_df
```

```python
sns.get_dataset_names()
```

```python
car_df.corr()
```

```python
plt.figure(figsize=(9,9))
sns.set_context('paper', font_scale=1.4)

# So we see that n previous accident is heavily correlated with accidents, while the
# insurance premium is not
crash_mx = car_df.corr()

# Create the heatmap, add annotations and a color map
sns.heatmap(crash_mx, 
            annot=True, 
            cmap='rocket'
           )
```
